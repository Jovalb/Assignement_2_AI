```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime as dt
from sklearn.model_selection import train_test_split
from sklearn import linear_model

df = pd.read_csv('NAS.csv')

df.head()
```

```python
df = df.dropna()
df = df.drop(['Open','High','Low','Adj Close','Volume'], axis='columns')
```

```python
df['Date'] = pd.to_datetime(df.Date, format='%Y-%m-%d')
```

```python
plt.figure(figsize=(16,8))
plt.plot(df['Date'],df['Close'], label='Close Price History')
```

```python
df['Date']=df['Date'].map(dt.datetime.toordinal)
df = df.sort_index(ascending=True, axis=0)
```

```python
future_days = 25
df['Prediction'] = df[['Close']].shift(-future_days)
```

```python
df
```

```python
x = np.array(df.drop(['Prediction','Close'], 1))[:-future_days]
```

```python
x
```

```python
y = np.array(df['Prediction'])[:-future_days]
print(y)
```

```python
print (x)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25)
```

```python
x_train
```

```python
learner = linear_model.LinearRegression()
```

```python
learner.fit(x_train,y_train)
```

```python
x_future = df.drop(['Prediction','Close'], 1)[:-future_days]
```

```python
x_future = x_future.tail(future_days)
```

```python
x_future = np.array(x_future)
x_future
```

```python
plt.xlabel('days',fontsize=16)
plt.ylabel('close price',fontsize=16)
#Visual Represention of linear equation with Linear Regression
plt.scatter(df.Date,df.Close,color='blue')
plt.plot(df.Date,learner.predict(df[['Date']]),color='red')
```
